```python
# Transpose the DataFrame
df_t = unemp.set_index('Country Name').T

# Reset the index to get 'Year' as a column
df_t = df_t.reset_index().rename(columns={'index': 'Year'})

# Drop the 'Country Name' column (which has been transposed)
df_t.columns.name = None
```

```python
# List of countries to filter
countries = ['Mexico', 'Honduras', 'Guatemala', 'El Salvador']

# Filtering the DataFrame
filtered_covid = covid[covid['location'].isin(countries)]
```

```python
filtered_covid = filtered_covid[['location', 'date', 'total_cases']]
```

```python
# Ensure that the 'date' column is in datetime format
filtered_covid['date'] = pd.to_datetime(filtered_covid['date'])

# Filter for rows where the date is the 1st of any month
filtered_covid_first_of_month = filtered_covid[filtered_covid['date'].dt.day == 1]

# Display the filtered dataframe
filtered_covid_first_of_month
```

```python
# Pivot the dataframe so that 'location' values become columns
reshaped_df = filtered_covid_first_of_month.pivot(index='date', columns='location', values='total_cases')

# Rename the columns to focus on Mexico, El Salvador, Guatemala, and Honduras
reshaped_df = reshaped_df[['Mexico', 'El Salvador', 'Guatemala', 'Honduras']].reset_index()

# Display the reshaped dataframe
reshaped_df
```

```python
# Filter the dataframe to only include rows where the date is before or equal to December 31, 2022
reshaped_df_until_dec_2022 = reshaped_df[reshaped_df['date'] <= '2022-12-31']

# Display the filtered dataframe
reshaped_df_until_dec_2022
```

```python
df2 = reshaped_df_until_dec_2022.copy()
```

```python
import pandas as pd

# Assuming enc1 is your DataFrame
# Strip whitespace from the column names
enc1.columns = enc1.columns.str.strip()

# Reshape the DataFrame using melt
enc1_melted = enc1.melt(id_vars=['Sectors', 'Year'],
                        var_name='Month',
                        value_name='Encounters')

# Map months to their corresponding numbers
month_map = {
    'Oct': '10', 'Nov': '11', 'Dec': '12', 'Jan': '01', 'Feb': '02', 'Mar': '03',
    'Apr': '04', 'May': '05', 'Jun': '06', 'Jul': '07', 'Aug': '08', 'Sep': '09'
}

# Replace the month names with their corresponding numbers
enc1_melted['Month'] = enc1_melted['Month'].map(month_map)

# Create a new date column in the format 'MM/DD/YYYY'
enc1_melted['Date'] = enc1_melted.apply(lambda row: f"01/{row['Month']}/{row['Year']}", axis=1)

# Drop unnecessary columns and reorder
enc1_final = enc1_melted[['Date', 'Sectors', 'Encounters']]
```

```python
df1 = enc1_final.copy()
```

```python
df1
```

```python
# Adding the new columns with placeholder data
df1['title42'] = None
df1['remain_in_mexico'] = None

# Convert 'Date' column to datetime for comparison
df1['Date'] = pd.to_datetime(df1['Date'], format='%d/%m/%Y')

# Define the date range for Title 42
start_date = datetime(2020, 3, 1)
end_date = datetime(2022, 5, 11)

# Update the 'title42' column based on the date range
df1['title42'] = df1['Date'].apply(lambda x: 1 if start_date <= x <= end_date else 0)

# Define the date range for Remain in Mexico
start_remain_in_mexico = datetime(2019, 1, 1)
end_remain_in_mexico = datetime(2022, 6, 30)

# Update the 'remain_in_mexico' column based on the date range
df1['remain_in_mexico'] = df1['Date'].apply(lambda x: 1 if start_remain_in_mexico <= x <= end_remain_in_mexico else 0)
```

```python
import pandas as pd

# For df1: Convert 'Date' column to datetime
df1['Date'] = pd.to_datetime(df1['Date'])

# For df2: Convert 'location' column to datetime (date column here is named 'location')
df2['date'] = pd.to_datetime(df2['date'])
df2.rename(columns={'date': 'Date'}, inplace=True)  # Rename to 'Date' to align with df1

# For df3: Since df3 only has 'Year', we'll convert that to the first day of the corresponding year
df3['Year'] = pd.to_datetime(df3['Year'], format='%Y')
```

```python
# Resample df3 unemployment data to monthly frequency and forward-fill
df3 = df3.set_index('Year').resample('MS').ffill().reset_index()  # 'MS' stands for Month Start
```

```python
# Merge df1 (encounters and policies) with df2 (COVID data for all four countries)
df_merged = pd.merge(df1, df2[['Date', 'Mexico', 'Guatemala', 'Honduras', 'El Salvador']], on='Date', how='left')

# Rename the columns to make them more descriptive
df_merged.rename(columns={
    'Mexico': 'Mexico_COVID_Cases',
    'Guatemala': 'Guatemala_COVID_Cases',
    'Honduras': 'Honduras_COVID_Cases',
    'El Salvador': 'ElSalvador_COVID_Cases'
}, inplace=True)
```

```python
# Merge unemployment data from df3 into df_merged
df_final = pd.merge(df_merged, df3[['Year', 'Mexico', 'Honduras', 'Guatemala', 'El Salvador']], left_on='Date', right_on='Year', how='left')

# Rename the unemployment columns to make them more descriptive
df_final.rename(columns={
    'Mexico': 'Mexico_Unemployment',
    'Honduras': 'Honduras_Unemployment',
    'Guatemala': 'Guatemala_Unemployment',
    'El Salvador': 'ElSalvador_Unemployment'
}, inplace=True)
```

```python
# Forward-fill missing data
df_final.fillna(method='ffill', inplace=True)

# Alternatively, you can interpolate missing values for a smoother transition
# df_final.interpolate(method='linear', inplace=True)
```

```python
# Fill missing COVID-19 case values with 0
df_final[['Mexico_COVID_Cases', 'Guatemala_COVID_Cases', 'Honduras_COVID_Cases', 'ElSalvador_COVID_Cases']] = df_final[['Mexico_COVID_Cases', 'Guatemala_COVID_Cases', 'Honduras_COVID_Cases', 'ElSalvador_COVID_Cases']].fillna(0)
```
